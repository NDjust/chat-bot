```python
import pandas as pd
```

```python
naver_shopping_data = []
steam_data = []

with open("./data/naver_shopping.txt", "r") as f:
    lines = f.readlines()

    for line in lines:
        naver_shopping_data.append(line.replace("\n", "").split("\t"))
        

with open("./data/steam.txt", "r") as f:
    lines = f.readlines()
    
    for line in lines:
        steam_data.append(line.replace("\n", "").split("\t"))
```
